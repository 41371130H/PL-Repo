```python
def format_date_with_weekday(date_str):
    """將 YYYYMMDD 格式的字串轉為 YYYY/MM/DD(週X)"""
    try:
        dt = datetime.datetime.strptime(str(date_str), '%Y%m%d')
        weekday_map = ["一", "二", "三", "四", "五", "六", "日"]
        return f"{dt.strftime('%Y/%m/%d')}({weekday_map[dt.weekday()]})"
    except (ValueError, TypeError):
        # 如果格式不對或不是字串，回傳原始值
        return date_str
```

```python
def query_todos(status, tag, due_date_option): # 將參數名改得更清晰
    """根據狀態、標籤和到期日篩選待辦事項"""
    global todos_data
    filtered_data = todos_data.copy()

    # 1.狀態篩選 (不變)
    if status != "全部":
        filtered_data = [t for t in filtered_data if t["狀態"] == status]

    # 2. 標籤篩選
    if tag != "全部":
        filtered_data = [t for t in filtered_data if t["標籤"] == tag]

    # 3.到期日篩選 (擴充邏輯)
    if due_date_option != "全部":
        today = datetime.date.today()
        temp_filtered = []
        for task in filtered_data:
            try:
                # 1. 使用 .get() 安全地獲取日期字串，如果沒有 "到期日" 欄位或為空，則返回空字串
                due_date_str = str(task.get("到期日", ""))

                if not due_date_str: # 如果日期為空，直接跳過此任務，進行下一個
                    continue

                # 2. 判斷字串中是否有 '(', 如果有才進行分割，否則直接使用原字串
                if '(' in due_date_str:
                    date_part = due_date_str.split('(')[0]
                else:
                    date_part = due_date_str # 直接使用，處理像 "2025/10/26" 這樣的格式

                # 3. 使用 strip() 去除可能存在的前後空白，然後再進行解析
                task_due_date = datetime.datetime.strptime(date_part.strip(), '%Y/%m/%d').date()

                if due_date_option == "今天到期" and task_due_date == today:
                    temp_filtered.append(task)
                elif due_date_option == "未來一週內到期" and today <= task_due_date < today + datetime.timedelta(days=7):
                    temp_filtered.append(task)
                elif due_date_option == "未來兩週內到期" and today <= task_due_date < today + datetime.timedelta(days=14):
                    temp_filtered.append(task)
                elif due_date_option == "未來一個月內到期" and today <= task_due_date < today + relativedelta(months=1):
                    temp_filtered.append(task)
                elif due_date_option == "已過期" and task_due_date < today:
                    temp_filtered.append(task)
                # "其他" 我們定義為「超過一個月後」的任務
                elif due_date_option == "未來一個月後" and task_due_date >= today + relativedelta(months=1):
                    temp_filtered.append(task)

            except (ValueError, TypeError):
                # 如果日期格式錯誤或為空，就忽略這個任務
                continue
        filtered_data = temp_filtered
    # 將篩選結果 (仍然是 list of dicts) 轉為 DataFrame 顯示
    # 注意：這裡的 filtered_data 裡的日期還是 YYYYMMDD
    # 我們交給 get_all_todos 去做最終的格式化顯示
    temp_df = pd.DataFrame(filtered_data)

    # 呼叫 get_all_todos 的邏輯來格式化，而不是自己重寫一遍
    if temp_df.empty:
        # 如果篩選後為空，返回一個帶有正確欄位的空 DataFrame
        df_full = pd.DataFrame(columns=["列號", "事項內容", "急迫性", "重要性", "到期日", "標籤", "狀態", "建立時間", "執行番茄數", "中斷次數", "手動結束"])
        return df_full, df_full # 返回兩個
    else:
        # 把篩選後的 DataFrame 傳給一個統一的格式化函數
        return format_and_split_df(temp_df)
```

```python
# 新增輔助函式，專門用來更新任務狀態和執行紀錄
def _update_task_status(row_index, new_status=None, stats=None):
    """更新指定列的狀態或執行統計 (同步更新記憶體與 Sheet)"""
    global todos_data # 確保我們能修改全域變數

    # [第一步] 更新記憶體快取 (todos_data)
    try:
        # 確保 row_index 在範圍內
        if row_index < 0 or row_index >= len(todos_data):
            print(f"❌ 更新記憶體快取失敗：索引 {row_index} 超出範圍。")
            return False

        if new_status:
            todos_data[row_index]["狀態"] = new_status

        if stats:
            # B4/B5: 累加邏輯
            # .get() 用 0 當作預設值，int() 處理 "0" 或 0
            old_pomodoros = int(todos_data[row_index].get("執行番茄數", 0))
            old_interrupts = int(todos_data[row_index].get("中斷次數", 0))
            old_manual_ends = int(todos_data[row_index].get("手動結束", 0))

            # 取得 "當前這次" 的統計
            current_pomodoros = stats.get("pomodoros_completed", 0)
            current_interrupts = stats.get("interruptions", 0)
            # B5: 將 "是"/"否" 轉換為 1/0
            is_manual_end = 1 if stats.get("manually_ended", "否") == "是" else 0

            # 累加
            todos_data[row_index]["執行番茄數"] = old_pomodoros + current_pomodoros
            todos_data[row_index]["中斷次數"] = old_interrupts + current_interrupts
            todos_data[row_index]["手動結束"] = old_manual_ends + is_manual_end

    except Exception as e:
        print(f"❌ 更新記憶體快取時發生錯誤：{e}")
        return False # 記憶體更新失敗，就不要更新 Sheet

    # [第二步] 更新 Google Sheet
    try:
        status_col = ws_todos.find("狀態").col
        sheet_row = row_index + 2 # 轉換為 gspread 的實際列號

        if new_status:
            ws_todos.update_cell(sheet_row, status_col, new_status)

        if stats:
            # 更新 H, I, J 欄 (使用累加後的值)
            exec_col = ws_todos.find("執行番茄數").col
            # 從剛更新完的 todos_data 讀取累加總數
            ws_todos.update_cell(sheet_row, exec_col, todos_data[row_index]["執行番茄數"])
            ws_todos.update_cell(sheet_row, exec_col + 1, todos_data[row_index]["中斷次數"])
            ws_todos.update_cell(sheet_row, exec_col + 2, todos_data[row_index]["手動結束"])

        return True
    except Exception as e:
        print(f"更新 Sheet 第 {row_index} 列時出錯: {e}")
        # 雖然 Sheet 更新失敗，但記憶體已更新，Gradio 介面仍會顯示
        # 可以在此加入重試邏輯，但目前保持簡單
        return False
```

```python
def export_data(format):
    """將目前的 todos_data 轉為 CSV 或 JSON 並回傳檔案路徑"""
    global todos_data
    df = pd.DataFrame(todos_data)
    if format == "CSV":
        filepath = "todos_export.csv"
        df.to_csv(filepath, index=False, encoding='utf-8-sig')
    else: # JSON
        filepath = "todos_export.json"
        # 使用 to_json 儲存為 json
        df.to_json(filepath, orient='records', indent=4, force_ascii=False)
    return filepath

def import_data(file_obj):
    """從上傳的檔案讀取資料，並覆蓋記憶體快取與整個 Google Sheet"""
    global todos_data
    if file_obj is None:
        return "⚠️ 沒有上傳檔案", *update_all_components()

    # ## 新增：在執行任何操作前，先深度備份目前的資料 ##
    backup_todos = copy.deepcopy(todos_data)

    try:
        filepath = file_obj.name
        if filepath.endswith(".csv"):
            df = pd.read_csv(filepath)
        elif filepath.endswith(".json"):
            df = pd.read_json(filepath, orient='records')
        else:
            return "❌ 不支援的檔案格式", *update_all_components()

        # [第一步] 更新記憶體快取
        # 確保匯入的資料欄位和我們需要的一致，避免錯誤
        expected_headers = ["事項內容", "急迫性", "重要性", "到期日", "標籤", "狀態", "建立時間", "執行番茄數", "中斷次數", "手動結束"]
        df = df.reindex(columns=expected_headers).fillna("") # 重新排序並填補缺失值
        todos_data = df.to_dict('records')

        # [第二步] 覆寫整個 Google Sheet
        ws_todos.clear()
        ws_todos.append_row(expected_headers, value_input_option="USER_ENTERED") # 寫入標頭
        if not df.empty:
            # 將 DataFrame 的值轉為 list of lists，再進行更新
            ws_todos.update('A2', df.values.tolist(), value_input_option="USER_ENTERED")

        return f"✅ 成功匯入並覆寫 {len(todos_data)} 筆紀錄", *update_all_components()
    except Exception as e:
        # ## 新增：如果發生任何錯誤，就從備份中還原 ##
        todos_data = backup_todos
        error_message = f"❌ 匯入失敗，已還原至原始狀態。錯誤：{e}"
        print(error_message) # 在後台印出詳細錯誤
        return error_message, *update_all_components()
```

```python
# ========== 功能 2：番茄鐘 ==========
def start_pomodoro(task_choice, num_pomodoros):
    """開始番茄鐘"""
    if not task_choice:
        # 返回4個值，對應4個 outputs
        main_df, ai_df = update_and_get_tables()
        return "請先選擇一個待辦事項", "🍅", main_df, ai_df, gr.update(visible=False)

    # 從 "【0】做簡報" 中提取出數字 0
    match = re.search(r"【(\d+)】", task_choice)
    if not match:
        main_df, ai_df = update_and_get_tables()
        return "選擇的任務格式錯誤", "🍅", main_df, ai_df, gr.update(visible=False)

    task_row_index = int(match.group(1))

    num_pomodoros = int(num_pomodoros)
    if num_pomodoros == 0:
        main_df, ai_df = update_and_get_tables()
        # 確保回傳 5 個值
        return "❌ 番茄數必須至少為 1", "🍅", main_df, ai_df, gr.update(visible=False)
    total_seconds = num_pomodoros * 25 * 60  # 每個番茄鐘 25 分鐘

    # 重置追蹤器並更新狀態為「進行中」
    timer_state.update({
        "running": True, "paused": False, "task": task_choice, "task_row": task_row_index,
        "pomodoros_completed": 0, "interruptions": 0, "manually_ended": "否",
        "remaining_time": total_seconds,
        "total_seconds_set": total_seconds # 儲存總時長
    })

    _update_task_status(task_row_index, new_status="進行中")


    timer_state["remaining_time"] = total_seconds

    # 立即更新一次介面
    main_df, ai_df = update_and_get_tables()
    yield f"計時開始：{task_choice}", f"{total_seconds // 60:02d}:{total_seconds % 60:02d}", main_df, ai_df, gr.update(visible=False)

    while timer_state["remaining_time"] > 0 and timer_state["running"]:
        time.sleep(1)
        if not timer_state["paused"]:
            timer_state["remaining_time"] -= 1

        if timer_state["paused"]:
            # 確保 task_choice 是最新的任務名稱
            status_message = f"已暫停：{timer_state['task']}"
        else:
            status_message = f"執行中：{timer_state['task']}"

        minutes = timer_state["remaining_time"] // 60
        seconds = timer_state["remaining_time"] % 60
        # 表格(todos)和按鈕(complete_btn)狀態不變，所以傳入 None


        # 根據 timer_state["paused"] 的狀態來決定要顯示的訊息
        if timer_state["paused"]:
            # 如果是暫停狀態，顯示「已暫停」
            status_message = f"已暫停：{timer_state['task']}"
        else:
            # 否則，顯示「執行中」
            status_message = f"執行中：{timer_state['task']}"

        # (請確保您有 5 個 outputs，對應到 start_btn.click 的 outputs)
        # outputs=[pomodoro_status, timer_display, todo_table_main, todo_table_ai, complete_btn]
        yield status_message, f"{minutes:02d}:{seconds:02d}", None, None, None

    if timer_state["running"]: # 正常結束
        timer_state["running"] = False
        timer_state["pomodoros_completed"] = num_pomodoros
        _update_task_status(task_row_index, stats=timer_state) # 寫入統計數據
        main_df, ai_df = update_and_get_tables()
        yield f"時間到！", "✅", main_df, ai_df, gr.update(visible=True, value="標示為已完成 👍") # 讓完成按鈕顯示

def pause_pomodoro():
    """暫停番茄鐘"""
    if timer_state["running"]:
        timer_state["paused"] = not timer_state["paused"]
        if timer_state["paused"]:
            timer_state["interruptions"] += 1 # 每次暫停時，中斷次數+1
            status_text = "已暫停"
        else:
            status_text = "執行中"

def end_pomodoro():
    """結束番茄鐘 (改為 generator 以提供即時反饋)"""
    if not timer_state["running"]:
        # 保持回傳 5 個值
        main_df, ai_df = update_and_get_tables()
        # 使用 yield 來停止
        yield "沒有正在運行的番茄鐘", "🍅", gr.update(visible=False), main_df, ai_df
        return # 必須有 return 來結束生成器

    original_task = timer_state["task"]
    task_row_to_update = timer_state["task_row"]

    # 1. 【增加提示訊息】
    # 立即回傳一個狀態，告知使用者系統正在運作
    # outputs=[pomodoro_status, timer_display, complete_btn, todo_table_main, todo_table_ai]
    # 我們只更新 pomodoro_status，其他傳入 None
    yield f"請稍等系統運作：{original_task}", None, None, None, None

    # [ 接下來是緩慢的操作 ]
    timer_state["running"] = False
    timer_state["manually_ended"] = "是"

    total_duration_set = timer_state.get("total_seconds_set", 0)
    remaining_time = timer_state["remaining_time"]
    elapsed_seconds = total_duration_set - remaining_time

    timer_state["pomodoros_completed"] = elapsed_seconds // (25 * 60)

    # 呼叫 gspread (緩慢)
    _update_task_status(task_row_to_update, stats=timer_state)
    # [ 緩慢的操作結束 ]

    # 清空狀態
    timer_state["task"] = ""
    timer_state["total_seconds_set"] = 0

    # [ 回傳最終結果 ]
    # 取得更新後的表格
    main_df, ai_df = update_and_get_tables()
    # 顯示完成按鈕，並更新表格
    yield f"已手動結束：{original_task}", "🍅", gr.update(visible=True, value="標示為已完成 👍"), main_df, ai_df
```

```python
# 新增函式，用於標示任務已完成
def mark_task_as_complete():
    """將計時器剛結束的任務標示為「已完成」"""
    task_row_to_mark = timer_state["task_row"] # 先抓出來
    if task_row_to_mark != -1:
        _update_task_status(task_row_to_mark, new_status="已完成")
        timer_state["task_row"] = -1 # 更新完再重置
        # 隱藏按鈕並更新表格
        main_df, ai_df = update_and_get_tables()
        return gr.update(visible=False), main_df, ai_df
```

```python
# ========== 功能 3：AI 回饋 ==========
def analyze_with_gemini():
    """使用 Gemini 分析待辦事項"""
    if not model:
        return "⚠️ Gemini 模型未設定，無法分析。"

    try:
        df = get_all_todos()
        if df.empty or len(df) < 1:
            return "⚠️ 沒有待辦事項可以分析。"

        # 準備 Prompt
        prompt = f"""
        這是我目前的待辦事項清單：
        {df.to_string(index=False)}

        請根據這份清單，以一個專業時間管理顧問的角度，用繁體中文提供我以下回饋：
        1.  **總體評估**：對我目前的任務量、重要性與急迫性分佈做一個簡短的總結。
        2.  **優先級建議**：根據「艾森豪矩陣」(重要又緊急 > 重要不緊急 > 緊急不重要 > 不重要不緊急)，建議我接下來應該優先處理哪 1-2 件事情，並說明原因。
        3.  **時間管理技巧**：給我一個具體的建議，幫助我更有效地完成這些任務（例如：是否有些任務可以批次處理？哪些任務適合在精神好的時候做？）。
        """

        response = model.generate_content(prompt, request_options={"timeout": 60})

        if hasattr(response, 'text') and response.text:
            advice = response.text
            # 清理 Markdown 字元
            advice = advice.replace("#", "").replace("*", "")
        else:
            advice = "⚠️ Gemini 分析完成，但沒有回傳文字內容。"

        # 寫入 Google Sheet
        tz = pytz.timezone("Asia/Taipei")
        timestamp = datetime.datetime.now(tz).strftime("%Y-%m-%d %H:%M:%S")
        ws_feedback.append_row([timestamp, advice], value_input_option="USER_ENTERED")

        return advice
    except Exception as e:
        return f"❌ AI 分析時發生錯誤：{e}"
```

```python
# ========== 功能 4：視覺化統計 ==========

def sync_and_update_analytics():
    """
    從 Google Sheet 同步最新資料，然後產生並回傳所有圖表。
    """
    print("🔄 正在從 Google Sheet 同步資料並更新圖表...")
    sync_from_sheet()  # 第一步：強制同步
    return update_analytics_tab() # 第二步：呼叫現有的函式來產生圖表

def create_status_pie_chart(data_df):
    """根據 DataFrame 產生狀態分佈的圓餅圖"""
    if data_df.empty or "狀態" not in data_df.columns:
        # 即使沒有資料，也回傳一個空的標題圖
        fig = px.pie(title="無資料可顯示")
        fig.update_layout(annotations=[dict(text="無資料", x=0.5, y=0.5, font_size=20, showarrow=False)])
        return fig

    status_counts = data_df["狀態"].value_counts().reset_index()
    status_counts.columns = ["狀態", "數量"]

    fig = px.pie(status_counts,
                 names="狀態",
                 values="數量",
                 title="任務狀態分佈",
                 color="狀態",
                 color_discrete_map={ # 自訂顏色
                     "已完成": "#63C54D",
                     "進行中": "#FFA500",
                     "未開始": "#FF6347"
                 })
    fig.update_layout(showlegend=True)
    return fig

def create_tag_bar_chart(data_df):
    """根據 DataFrame 產生標籤分佈的長條圖"""
    if data_df.empty or "標籤" not in data_df.columns:
        fig = px.bar(title="無資料可顯示")
        fig.update_layout(annotations=[dict(text="無資料", x=0.5, y=0.5, font_size=20, showarrow=False)])
        return fig

    tag_counts = data_df["標籤"].value_counts().reset_index()
    tag_counts.columns = ["標籤", "數量"]

    fig = px.bar(tag_counts,
                 x="標籤",
                 y="數量",
                 title="任務標籤分佈",
                 color="標籤",
                 text_auto=True) # 在長條上顯示數字
    return fig

def update_analytics_tab():
    """準備資料並回傳所有圖表 (給 Gradio 呼叫)"""
    global todos_data
    if not todos_data:
        # 如果沒資料，回傳空的圖表
        empty_df = pd.DataFrame(columns=["狀態", "標籤"])
        return create_status_pie_chart(empty_df), create_tag_bar_chart(empty_df)

    # 使用全域變數
    df = pd.DataFrame(todos_data)

    # 確保 '狀態' 和 '標籤' 是字串，避免 Plotly 處理 NaNs 或數字時出錯
    df["狀態"] = df["狀態"].astype(str).replace("nan", "未分類")
    df["標籤"] = df["標籤"].astype(str).replace("nan", "未分類")

    # 產生圖表
    fig_status = create_status_pie_chart(df)
    fig_tag = create_tag_bar_chart(df)

    return fig_status, fig_tag

def update_all_components():
    """
    一個集中的函數，用於獲取所有更新的資料和圖表。
    它將回傳 7 個項目：2 個表格，2 個圖表，3 個篩選器重置狀態。
    """
    # 1. 獲取更新後的表格
    main_df, ai_df = update_and_get_tables()

    # 2. 獲取更新後的圖表
    fig_status, fig_tag = update_analytics_tab()

    # 3. 準備重置篩選器
    reset_status_filter = gr.update(value="全部")
    reset_tag_filter = gr.update(value="全部")
    reset_due_date_filter = gr.update(value="全部")

    return main_df, ai_df, fig_status, fig_tag, reset_status_filter, reset_tag_filter, reset_due_date_filter
```
