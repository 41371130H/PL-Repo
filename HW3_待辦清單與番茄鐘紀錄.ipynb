```python
def format_and_split_df(input_df):
    """æ¥æ”¶ä¸€å€‹ DataFrameï¼Œæ ¼å¼åŒ–å¾Œå›å‚³ä¸»è¡¨æ ¼å’Œ AI è¡¨æ ¼çš„ç‰ˆæœ¬"""
    # æ ¼å¼åŒ–æ—¥æœŸæ¬„ä½
    if "åˆ°æœŸæ—¥" in input_df.columns:
        input_df["åˆ°æœŸæ—¥"] = input_df["åˆ°æœŸæ—¥"].apply(format_date_with_weekday)

    # ç¢ºä¿æ¬„ä½é½Šå…¨ä¸¦åŠ ä¸Šåˆ—è™Ÿ
    all_headers = ["äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹", "å»ºç«‹æ™‚é–“", "åŸ·è¡Œç•ªèŒ„æ•¸", "ä¸­æ–·æ¬¡æ•¸", "æ‰‹å‹•çµæŸ"]
    df_full = input_df.reindex(columns=all_headers).fillna("")
    df_full_with_index = df_full.reset_index(drop=True).reset_index().rename(columns={"index": "åˆ—è™Ÿ"})

    # æº–å‚™ä¸»åˆ†é çš„æ‘˜è¦ç‰ˆæœ¬
    main_headers = ["åˆ—è™Ÿ", "äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹", "å»ºç«‹æ™‚é–“"]
    df_main = df_full_with_index[main_headers]

    return df_main, df_full_with_index

def update_and_get_tables():
    """å¾å…¨åŸŸè®Šæ•¸ todos_data å–å¾—è³‡æ–™ä¸¦æ ¼å¼åŒ–æˆå…©å€‹è¡¨æ ¼"""
    df = pd.DataFrame(todos_data)
    if df.empty:
        # å»ºç«‹å…©å€‹ç©ºçš„ DataFrame
        main_headers = ["åˆ—è™Ÿ", "äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹", "å»ºç«‹æ™‚é–“"]
        ai_headers = ["åˆ—è™Ÿ"] + list(pd.DataFrame(columns=todos_data[0].keys()).columns) if todos_data else []
        return pd.DataFrame(columns=main_headers), pd.DataFrame(columns=ai_headers)

    return format_and_split_df(df)
```

```python
# ========== åŠŸèƒ½ 1ï¼šé¡¯ç¤º / æ–°å¢ / åˆªé™¤å¾…è¾¦äº‹é … ==========
def get_all_todos():
    """å¾å…¨åŸŸè®Šæ•¸ todos_data (å¿«å–) è®€å–æ‰€æœ‰å¾…è¾¦äº‹é …ä¸¦æ ¼å¼åŒ–ä»¥ä¾›é¡¯ç¤º"""
    global todos_data
    if not todos_data:
        return pd.DataFrame(columns=["åˆ—è™Ÿ", "äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹",
                                    "å»ºç«‹æ™‚é–“", "åŸ·è¡Œç•ªèŒ„æ•¸", "ä¸­æ–·æ¬¡æ•¸", "æ‰‹å‹•çµæŸ"])

    df = pd.DataFrame(todos_data)
    if df.empty:
        return df.reindex(columns=["åˆ—è™Ÿ", "äº‹é …å…§å®¹", ...]) # ç¢ºä¿ç©ºè¡¨æ ¼ä¹Ÿæœ‰æ¬„ä½

    # ## æ–°å¢ï¼šæ ¼å¼åŒ–ã€Œåˆ°æœŸæ—¥ã€æ¬„ä½ ##
    if "åˆ°æœŸæ—¥" in df.columns:
        df["åˆ°æœŸæ—¥"] = df["åˆ°æœŸæ—¥"].apply(format_date_with_weekday)

    # ç¢ºä¿æ¬„ä½é †åº
    expected_headers = ["äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹", "å»ºç«‹æ™‚é–“", "åŸ·è¡Œç•ªèŒ„æ•¸", "ä¸­æ–·æ¬¡æ•¸", "æ‰‹å‹•çµæŸ"]
    df = df.reindex(columns=expected_headers)

    df_with_index = df.reset_index(drop=True).reset_index().rename(columns={"index": "åˆ—è™Ÿ"})
    return df_with_index

def add_todo(content, urgency, importance, due_date, tag, status):
    """æ–°å¢äº‹é …åˆ°è¨˜æ†¶é«”å¿«å–å’Œ Google Sheet"""
    global todos_data
    if not content:
      return "âŒ äº‹é …å…§å®¹ä¸å¯ç©ºç™½", *update_all_components()

    try:
      tz = pytz.timezone("Asia/Taipei")
      timestamp = datetime.datetime.now(tz).strftime("%Y-%m-%d %H:%M")
      new_todo_dict = {
          "äº‹é …å…§å®¹": content, "æ€¥è¿«æ€§": urgency, "é‡è¦æ€§": importance, "åˆ°æœŸæ—¥": due_date,
          "æ¨™ç±¤": tag, "ç‹€æ…‹": status, "å»ºç«‹æ™‚é–“": timestamp,
          "åŸ·è¡Œç•ªèŒ„æ•¸": 0, "ä¸­æ–·æ¬¡æ•¸": 0, "æ‰‹å‹•çµæŸ": 0
      }
      # ## [ç¬¬ä¸€æ­¥] æ›´æ–°è¨˜æ†¶é«”å¿«å– ##
      todos_data.append(new_todo_dict)

      # [ç¬¬äºŒæ­¥] æº–å‚™å¯«å…¥ Sheet çš„è³‡æ–™ï¼Œä¸¦åœ¨æ­¤æ™‚æ ¼å¼åŒ–æ—¥æœŸ
      row_to_write = list(new_todo_dict.values())
      # æ‰¾åˆ° "åˆ°æœŸæ—¥" åœ¨ dict ä¸­çš„ä½ç½® (é€šå¸¸æ˜¯ç¬¬4å€‹ï¼Œç´¢å¼•ç‚º3)
      date_index = list(new_todo_dict.keys()).index("åˆ°æœŸæ—¥")
      # ## æ–°å¢ï¼šåªåœ¨å¯«å…¥ Sheet å‰æ ¼å¼åŒ– ##
      row_to_write[date_index] = format_date_with_weekday(due_date)

      ws_todos.append_row(row_to_write, value_input_option="USER_ENTERED")

      return f"âœ… å·²æ–°å¢äº‹é …ï¼š{content}", *update_all_components()
    except Exception as e:
      # å¦‚æœå¯«å…¥ Sheet å¤±æ•—ï¼Œæœ€å¥½ä¹ŸæŠŠå‰›åŠ å…¥å¿«å–çš„è³‡æ–™ç§»é™¤ï¼Œä¿æŒä¸€è‡´æ€§
      todos_data.pop()
      return f"âŒ æ–°å¢å¤±æ•—ï¼š{e}", *update_all_components()

def record_selection(evt: gr.SelectData):
    """ç•¶ä½¿ç”¨è€…é»æ“Šè¡¨æ ¼æ™‚ï¼Œè¨˜éŒ„ä¸‹è¢«é¸ä¸­çš„åˆ—è™Ÿ"""
    if evt.index: # evt.index æ˜¯ä¸€å€‹ (row, col) çš„ tuple
        return evt.index[0] # æˆ‘å€‘åªéœ€è¦åˆ—è™Ÿ (row index)
    return None

def delete_selected_todo(selected_index):
    """æ ¹æ“š selected_index åˆªé™¤è¨˜æ†¶é«”å¿«å–å’Œ Google Sheet ä¸­çš„å¾…è¾¦äº‹é …"""
    global todos_data
    if selected_index is None or selected_index < 0 or selected_index >= len(todos_data):
      return "âš ï¸ è«‹å…ˆå¾ä¸Šæ–¹è¡¨æ ¼é»é¸ä¸€å€‹é …ç›®", *update_all_components()

    try:
      selected_index = int(selected_index)
      # ## [ç¬¬ä¸€æ­¥] å¾è¨˜æ†¶é«”å¿«å–ç§»é™¤ ##
      deleted_task = todos_data.pop(selected_index)

      # ## [ç¬¬äºŒæ­¥] å¾ Google Sheet åˆªé™¤ ##
      # Google Sheets çš„åˆ—è™Ÿæ˜¯å¾ 1 é–‹å§‹ï¼Œä¸”æ¨™é¡Œä½”ç¬¬ 1 åˆ—
      # æ‰€ä»¥ DataFrame çš„ã€Œåˆ—è™Ÿ nã€å°æ‡‰åˆ° Sheet çš„ç¬¬ n+2 åˆ—
      ws_todos.delete_rows(selected_index + 2)

      return f"ğŸ—‘ï¸ å·²åˆªé™¤äº‹é …ï¼š{deleted_task['äº‹é …å…§å®¹']}", *update_all_components()
    except Exception as e:
      # å¦‚æœæ“ä½œå¤±æ•—ï¼Œæœ€å¥½é‡æ–°åŒæ­¥ä¸€æ¬¡ï¼Œç¢ºä¿è³‡æ–™ä¸€è‡´
      sync_from_sheet()
      return f"âŒ åˆªé™¤å¤±æ•—ï¼š{e}", *update_all_components()

def update_selected_todo_status(selected_index):
    """å°‡é¸å®šçš„å¾…è¾¦äº‹é …ç‹€æ…‹æ”¹ç‚ºã€Œå·²å®Œæˆã€(åŒæ­¥æ›´æ–°è¨˜æ†¶é«”èˆ‡ Sheet)"""
    global todos_data
    if selected_index is None or selected_index < 0 or selected_index >= len(todos_data):
      return "âš ï¸ è«‹å…ˆå¾ä¸Šæ–¹è¡¨æ ¼é»é¸ä¸€å€‹é …ç›®", *update_all_components()

    try:
      selected_index = int(selected_index)
      # ## [ç¬¬ä¸€æ­¥] æ›´æ–°è¨˜æ†¶é«”å¿«å– ##
      todos_data[selected_index]["ç‹€æ…‹"] = "å·²å®Œæˆ"

      # ## [ç¬¬äºŒæ­¥] æ›´æ–° Google Sheet ##
      sheet_row = selected_index + 2
      # æ‰¾åˆ°ã€Œç‹€æ…‹ã€æ˜¯ç¬¬å¹¾æ¬„ (Fæ¬„ = 6)
      # ç‚ºäº†ç©©å®šæ€§ï¼Œå¯ä»¥ç¡¬ç·¨ç¢¼æˆ–ç”¨ ws_todos.find("ç‹€æ…‹").col
      status_col_index = 6
      ws_todos.update_cell(sheet_row, status_col_index, "å·²å®Œæˆ")

      return f"âœ… å·²æ›´æ–°ç‹€æ…‹", *update_all_components()
    except Exception as e:
      sync_from_sheet()
      return f"âŒ æ›´æ–°ç‹€æ…‹å¤±æ•—ï¼š{e}", *update_all_components()

# åŒæ¨£åœ°ï¼Œä¹Ÿéœ€è¦ä¿®æ”¹ç•ªèŒ„é˜ä½¿ç”¨çš„ `_update_task_status()` å‡½æ•¸ï¼Œè®“å®ƒåŒæ™‚æ›´æ–°è¨˜æ†¶é«”å’Œ Sheetã€‚
```

```python
def format_date_with_weekday(date_str):
    """å°‡ YYYYMMDD æ ¼å¼çš„å­—ä¸²è½‰ç‚º YYYY/MM/DD(é€±X)"""
    try:
        dt = datetime.datetime.strptime(str(date_str), '%Y%m%d')
        weekday_map = ["ä¸€", "äºŒ", "ä¸‰", "å››", "äº”", "å…­", "æ—¥"]
        return f"{dt.strftime('%Y/%m/%d')}({weekday_map[dt.weekday()]})"
    except (ValueError, TypeError):
        # å¦‚æœæ ¼å¼ä¸å°æˆ–ä¸æ˜¯å­—ä¸²ï¼Œå›å‚³åŸå§‹å€¼
        return date_str
```

```python
def query_todos(status, tag, due_date_option): # å°‡åƒæ•¸åæ”¹å¾—æ›´æ¸…æ™°
    """æ ¹æ“šç‹€æ…‹ã€æ¨™ç±¤å’Œåˆ°æœŸæ—¥ç¯©é¸å¾…è¾¦äº‹é …"""
    global todos_data
    filtered_data = todos_data.copy()

    # 1.ç‹€æ…‹ç¯©é¸ (ä¸è®Š)
    if status != "å…¨éƒ¨":
        filtered_data = [t for t in filtered_data if t["ç‹€æ…‹"] == status]

    # 2. æ¨™ç±¤ç¯©é¸
    if tag != "å…¨éƒ¨":
        filtered_data = [t for t in filtered_data if t["æ¨™ç±¤"] == tag]

    # 3.åˆ°æœŸæ—¥ç¯©é¸ (æ“´å……é‚è¼¯)
    if due_date_option != "å…¨éƒ¨":
        today = datetime.date.today()
        temp_filtered = []
        for task in filtered_data:
            try:
                # 1. ä½¿ç”¨ .get() å®‰å…¨åœ°ç²å–æ—¥æœŸå­—ä¸²ï¼Œå¦‚æœæ²’æœ‰ "åˆ°æœŸæ—¥" æ¬„ä½æˆ–ç‚ºç©ºï¼Œå‰‡è¿”å›ç©ºå­—ä¸²
                due_date_str = str(task.get("åˆ°æœŸæ—¥", ""))

                if not due_date_str: # å¦‚æœæ—¥æœŸç‚ºç©ºï¼Œç›´æ¥è·³éæ­¤ä»»å‹™ï¼Œé€²è¡Œä¸‹ä¸€å€‹
                    continue

                # 2. åˆ¤æ–·å­—ä¸²ä¸­æ˜¯å¦æœ‰ '(', å¦‚æœæœ‰æ‰é€²è¡Œåˆ†å‰²ï¼Œå¦å‰‡ç›´æ¥ä½¿ç”¨åŸå­—ä¸²
                if '(' in due_date_str:
                    date_part = due_date_str.split('(')[0]
                else:
                    date_part = due_date_str # ç›´æ¥ä½¿ç”¨ï¼Œè™•ç†åƒ "2025/10/26" é€™æ¨£çš„æ ¼å¼

                # 3. ä½¿ç”¨ strip() å»é™¤å¯èƒ½å­˜åœ¨çš„å‰å¾Œç©ºç™½ï¼Œç„¶å¾Œå†é€²è¡Œè§£æ
                task_due_date = datetime.datetime.strptime(date_part.strip(), '%Y/%m/%d').date()

                if due_date_option == "ä»Šå¤©åˆ°æœŸ" and task_due_date == today:
                    temp_filtered.append(task)
                elif due_date_option == "æœªä¾†ä¸€é€±å…§åˆ°æœŸ" and today <= task_due_date < today + datetime.timedelta(days=7):
                    temp_filtered.append(task)
                elif due_date_option == "æœªä¾†å…©é€±å…§åˆ°æœŸ" and today <= task_due_date < today + datetime.timedelta(days=14):
                    temp_filtered.append(task)
                elif due_date_option == "æœªä¾†ä¸€å€‹æœˆå…§åˆ°æœŸ" and today <= task_due_date < today + relativedelta(months=1):
                    temp_filtered.append(task)
                elif due_date_option == "å·²éæœŸ" and task_due_date < today:
                    temp_filtered.append(task)
                # "å…¶ä»–" æˆ‘å€‘å®šç¾©ç‚ºã€Œè¶…éä¸€å€‹æœˆå¾Œã€çš„ä»»å‹™
                elif due_date_option == "æœªä¾†ä¸€å€‹æœˆå¾Œ" and task_due_date >= today + relativedelta(months=1):
                    temp_filtered.append(task)

            except (ValueError, TypeError):
                # å¦‚æœæ—¥æœŸæ ¼å¼éŒ¯èª¤æˆ–ç‚ºç©ºï¼Œå°±å¿½ç•¥é€™å€‹ä»»å‹™
                continue
        filtered_data = temp_filtered
    # å°‡ç¯©é¸çµæœ (ä»ç„¶æ˜¯ list of dicts) è½‰ç‚º DataFrame é¡¯ç¤º
    # æ³¨æ„ï¼šé€™è£¡çš„ filtered_data è£¡çš„æ—¥æœŸé‚„æ˜¯ YYYYMMDD
    # æˆ‘å€‘äº¤çµ¦ get_all_todos å»åšæœ€çµ‚çš„æ ¼å¼åŒ–é¡¯ç¤º
    temp_df = pd.DataFrame(filtered_data)

    # å‘¼å« get_all_todos çš„é‚è¼¯ä¾†æ ¼å¼åŒ–ï¼Œè€Œä¸æ˜¯è‡ªå·±é‡å¯«ä¸€é
    if temp_df.empty:
        # å¦‚æœç¯©é¸å¾Œç‚ºç©ºï¼Œè¿”å›ä¸€å€‹å¸¶æœ‰æ­£ç¢ºæ¬„ä½çš„ç©º DataFrame
        df_full = pd.DataFrame(columns=["åˆ—è™Ÿ", "äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹", "å»ºç«‹æ™‚é–“", "åŸ·è¡Œç•ªèŒ„æ•¸", "ä¸­æ–·æ¬¡æ•¸", "æ‰‹å‹•çµæŸ"])
        return df_full, df_full # è¿”å›å…©å€‹
    else:
        # æŠŠç¯©é¸å¾Œçš„ DataFrame å‚³çµ¦ä¸€å€‹çµ±ä¸€çš„æ ¼å¼åŒ–å‡½æ•¸
        return format_and_split_df(temp_df)
```

```python
# æ–°å¢è¼”åŠ©å‡½å¼ï¼Œå°ˆé–€ç”¨ä¾†æ›´æ–°ä»»å‹™ç‹€æ…‹å’ŒåŸ·è¡Œç´€éŒ„
def _update_task_status(row_index, new_status=None, stats=None):
    """æ›´æ–°æŒ‡å®šåˆ—çš„ç‹€æ…‹æˆ–åŸ·è¡Œçµ±è¨ˆ (åŒæ­¥æ›´æ–°è¨˜æ†¶é«”èˆ‡ Sheet)"""
    global todos_data # ç¢ºä¿æˆ‘å€‘èƒ½ä¿®æ”¹å…¨åŸŸè®Šæ•¸

    # [ç¬¬ä¸€æ­¥] æ›´æ–°è¨˜æ†¶é«”å¿«å– (todos_data)
    try:
        # ç¢ºä¿ row_index åœ¨ç¯„åœå…§
        if row_index < 0 or row_index >= len(todos_data):
            print(f"âŒ æ›´æ–°è¨˜æ†¶é«”å¿«å–å¤±æ•—ï¼šç´¢å¼• {row_index} è¶…å‡ºç¯„åœã€‚")
            return False

        if new_status:
            todos_data[row_index]["ç‹€æ…‹"] = new_status

        if stats:
            # B4/B5: ç´¯åŠ é‚è¼¯
            # .get() ç”¨ 0 ç•¶ä½œé è¨­å€¼ï¼Œint() è™•ç† "0" æˆ– 0
            old_pomodoros = int(todos_data[row_index].get("åŸ·è¡Œç•ªèŒ„æ•¸", 0))
            old_interrupts = int(todos_data[row_index].get("ä¸­æ–·æ¬¡æ•¸", 0))
            old_manual_ends = int(todos_data[row_index].get("æ‰‹å‹•çµæŸ", 0))

            # å–å¾— "ç•¶å‰é€™æ¬¡" çš„çµ±è¨ˆ
            current_pomodoros = stats.get("pomodoros_completed", 0)
            current_interrupts = stats.get("interruptions", 0)
            # B5: å°‡ "æ˜¯"/"å¦" è½‰æ›ç‚º 1/0
            is_manual_end = 1 if stats.get("manually_ended", "å¦") == "æ˜¯" else 0

            # ç´¯åŠ 
            todos_data[row_index]["åŸ·è¡Œç•ªèŒ„æ•¸"] = old_pomodoros + current_pomodoros
            todos_data[row_index]["ä¸­æ–·æ¬¡æ•¸"] = old_interrupts + current_interrupts
            todos_data[row_index]["æ‰‹å‹•çµæŸ"] = old_manual_ends + is_manual_end

    except Exception as e:
        print(f"âŒ æ›´æ–°è¨˜æ†¶é«”å¿«å–æ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
        return False # è¨˜æ†¶é«”æ›´æ–°å¤±æ•—ï¼Œå°±ä¸è¦æ›´æ–° Sheet

    # [ç¬¬äºŒæ­¥] æ›´æ–° Google Sheet
    try:
        status_col = ws_todos.find("ç‹€æ…‹").col
        sheet_row = row_index + 2 # è½‰æ›ç‚º gspread çš„å¯¦éš›åˆ—è™Ÿ

        if new_status:
            ws_todos.update_cell(sheet_row, status_col, new_status)

        if stats:
            # æ›´æ–° H, I, J æ¬„ (ä½¿ç”¨ç´¯åŠ å¾Œçš„å€¼)
            exec_col = ws_todos.find("åŸ·è¡Œç•ªèŒ„æ•¸").col
            # å¾å‰›æ›´æ–°å®Œçš„ todos_data è®€å–ç´¯åŠ ç¸½æ•¸
            ws_todos.update_cell(sheet_row, exec_col, todos_data[row_index]["åŸ·è¡Œç•ªèŒ„æ•¸"])
            ws_todos.update_cell(sheet_row, exec_col + 1, todos_data[row_index]["ä¸­æ–·æ¬¡æ•¸"])
            ws_todos.update_cell(sheet_row, exec_col + 2, todos_data[row_index]["æ‰‹å‹•çµæŸ"])

        return True
    except Exception as e:
        print(f"æ›´æ–° Sheet ç¬¬ {row_index} åˆ—æ™‚å‡ºéŒ¯: {e}")
        # é›–ç„¶ Sheet æ›´æ–°å¤±æ•—ï¼Œä½†è¨˜æ†¶é«”å·²æ›´æ–°ï¼ŒGradio ä»‹é¢ä»æœƒé¡¯ç¤º
        # å¯ä»¥åœ¨æ­¤åŠ å…¥é‡è©¦é‚è¼¯ï¼Œä½†ç›®å‰ä¿æŒç°¡å–®
        return False
```

```python
def export_data(format):
    """å°‡ç›®å‰çš„ todos_data è½‰ç‚º CSV æˆ– JSON ä¸¦å›å‚³æª”æ¡ˆè·¯å¾‘"""
    global todos_data
    df = pd.DataFrame(todos_data)
    if format == "CSV":
        filepath = "todos_export.csv"
        df.to_csv(filepath, index=False, encoding='utf-8-sig')
    else: # JSON
        filepath = "todos_export.json"
        # ä½¿ç”¨ to_json å„²å­˜ç‚º json
        df.to_json(filepath, orient='records', indent=4, force_ascii=False)
    return filepath

def import_data(file_obj):
    """å¾ä¸Šå‚³çš„æª”æ¡ˆè®€å–è³‡æ–™ï¼Œä¸¦è¦†è“‹è¨˜æ†¶é«”å¿«å–èˆ‡æ•´å€‹ Google Sheet"""
    global todos_data
    if file_obj is None:
        return "âš ï¸ æ²’æœ‰ä¸Šå‚³æª”æ¡ˆ", *update_all_components()

    # ## æ–°å¢ï¼šåœ¨åŸ·è¡Œä»»ä½•æ“ä½œå‰ï¼Œå…ˆæ·±åº¦å‚™ä»½ç›®å‰çš„è³‡æ–™ ##
    backup_todos = copy.deepcopy(todos_data)

    try:
        filepath = file_obj.name
        if filepath.endswith(".csv"):
            df = pd.read_csv(filepath)
        elif filepath.endswith(".json"):
            df = pd.read_json(filepath, orient='records')
        else:
            return "âŒ ä¸æ”¯æ´çš„æª”æ¡ˆæ ¼å¼", *update_all_components()

        # [ç¬¬ä¸€æ­¥] æ›´æ–°è¨˜æ†¶é«”å¿«å–
        # ç¢ºä¿åŒ¯å…¥çš„è³‡æ–™æ¬„ä½å’Œæˆ‘å€‘éœ€è¦çš„ä¸€è‡´ï¼Œé¿å…éŒ¯èª¤
        expected_headers = ["äº‹é …å…§å®¹", "æ€¥è¿«æ€§", "é‡è¦æ€§", "åˆ°æœŸæ—¥", "æ¨™ç±¤", "ç‹€æ…‹", "å»ºç«‹æ™‚é–“", "åŸ·è¡Œç•ªèŒ„æ•¸", "ä¸­æ–·æ¬¡æ•¸", "æ‰‹å‹•çµæŸ"]
        df = df.reindex(columns=expected_headers).fillna("") # é‡æ–°æ’åºä¸¦å¡«è£œç¼ºå¤±å€¼
        todos_data = df.to_dict('records')

        # [ç¬¬äºŒæ­¥] è¦†å¯«æ•´å€‹ Google Sheet
        ws_todos.clear()
        ws_todos.append_row(expected_headers, value_input_option="USER_ENTERED") # å¯«å…¥æ¨™é ­
        if not df.empty:
            # å°‡ DataFrame çš„å€¼è½‰ç‚º list of listsï¼Œå†é€²è¡Œæ›´æ–°
            ws_todos.update('A2', df.values.tolist(), value_input_option="USER_ENTERED")

        return f"âœ… æˆåŠŸåŒ¯å…¥ä¸¦è¦†å¯« {len(todos_data)} ç­†ç´€éŒ„", *update_all_components()
    except Exception as e:
        # ## æ–°å¢ï¼šå¦‚æœç™¼ç”Ÿä»»ä½•éŒ¯èª¤ï¼Œå°±å¾å‚™ä»½ä¸­é‚„åŸ ##
        todos_data = backup_todos
        error_message = f"âŒ åŒ¯å…¥å¤±æ•—ï¼Œå·²é‚„åŸè‡³åŸå§‹ç‹€æ…‹ã€‚éŒ¯èª¤ï¼š{e}"
        print(error_message) # åœ¨å¾Œå°å°å‡ºè©³ç´°éŒ¯èª¤
        return error_message, *update_all_components()
```

```python
# ========== åŠŸèƒ½ 2ï¼šç•ªèŒ„é˜ ==========
def start_pomodoro(task_choice, num_pomodoros):
    """é–‹å§‹ç•ªèŒ„é˜"""
    if not task_choice:
        # è¿”å›4å€‹å€¼ï¼Œå°æ‡‰4å€‹ outputs
        main_df, ai_df = update_and_get_tables()
        return "è«‹å…ˆé¸æ“‡ä¸€å€‹å¾…è¾¦äº‹é …", "ğŸ…", main_df, ai_df, gr.update(visible=False)

    # å¾ "ã€0ã€‘åšç°¡å ±" ä¸­æå–å‡ºæ•¸å­— 0
    match = re.search(r"ã€(\d+)ã€‘", task_choice)
    if not match:
        main_df, ai_df = update_and_get_tables()
        return "é¸æ“‡çš„ä»»å‹™æ ¼å¼éŒ¯èª¤", "ğŸ…", main_df, ai_df, gr.update(visible=False)

    task_row_index = int(match.group(1))

    num_pomodoros = int(num_pomodoros)
    if num_pomodoros == 0:
        main_df, ai_df = update_and_get_tables()
        # ç¢ºä¿å›å‚³ 5 å€‹å€¼
        return "âŒ ç•ªèŒ„æ•¸å¿…é ˆè‡³å°‘ç‚º 1", "ğŸ…", main_df, ai_df, gr.update(visible=False)
    total_seconds = num_pomodoros * 25 * 60  # æ¯å€‹ç•ªèŒ„é˜ 25 åˆ†é˜

    # é‡ç½®è¿½è¹¤å™¨ä¸¦æ›´æ–°ç‹€æ…‹ç‚ºã€Œé€²è¡Œä¸­ã€
    timer_state.update({
        "running": True, "paused": False, "task": task_choice, "task_row": task_row_index,
        "pomodoros_completed": 0, "interruptions": 0, "manually_ended": "å¦",
        "remaining_time": total_seconds,
        "total_seconds_set": total_seconds # å„²å­˜ç¸½æ™‚é•·
    })

    _update_task_status(task_row_index, new_status="é€²è¡Œä¸­")


    timer_state["remaining_time"] = total_seconds

    # ç«‹å³æ›´æ–°ä¸€æ¬¡ä»‹é¢
    main_df, ai_df = update_and_get_tables()
    yield f"è¨ˆæ™‚é–‹å§‹ï¼š{task_choice}", f"{total_seconds // 60:02d}:{total_seconds % 60:02d}", main_df, ai_df, gr.update(visible=False)

    while timer_state["remaining_time"] > 0 and timer_state["running"]:
        time.sleep(1)
        if not timer_state["paused"]:
            timer_state["remaining_time"] -= 1

        if timer_state["paused"]:
            # ç¢ºä¿ task_choice æ˜¯æœ€æ–°çš„ä»»å‹™åç¨±
            status_message = f"å·²æš«åœï¼š{timer_state['task']}"
        else:
            status_message = f"åŸ·è¡Œä¸­ï¼š{timer_state['task']}"

        minutes = timer_state["remaining_time"] // 60
        seconds = timer_state["remaining_time"] % 60
        # è¡¨æ ¼(todos)å’ŒæŒ‰éˆ•(complete_btn)ç‹€æ…‹ä¸è®Šï¼Œæ‰€ä»¥å‚³å…¥ None


        # æ ¹æ“š timer_state["paused"] çš„ç‹€æ…‹ä¾†æ±ºå®šè¦é¡¯ç¤ºçš„è¨Šæ¯
        if timer_state["paused"]:
            # å¦‚æœæ˜¯æš«åœç‹€æ…‹ï¼Œé¡¯ç¤ºã€Œå·²æš«åœã€
            status_message = f"å·²æš«åœï¼š{timer_state['task']}"
        else:
            # å¦å‰‡ï¼Œé¡¯ç¤ºã€ŒåŸ·è¡Œä¸­ã€
            status_message = f"åŸ·è¡Œä¸­ï¼š{timer_state['task']}"

        # (è«‹ç¢ºä¿æ‚¨æœ‰ 5 å€‹ outputsï¼Œå°æ‡‰åˆ° start_btn.click çš„ outputs)
        # outputs=[pomodoro_status, timer_display, todo_table_main, todo_table_ai, complete_btn]
        yield status_message, f"{minutes:02d}:{seconds:02d}", None, None, None

    if timer_state["running"]: # æ­£å¸¸çµæŸ
        timer_state["running"] = False
        timer_state["pomodoros_completed"] = num_pomodoros
        _update_task_status(task_row_index, stats=timer_state) # å¯«å…¥çµ±è¨ˆæ•¸æ“š
        main_df, ai_df = update_and_get_tables()
        yield f"æ™‚é–“åˆ°ï¼", "âœ…", main_df, ai_df, gr.update(visible=True, value="æ¨™ç¤ºç‚ºå·²å®Œæˆ ğŸ‘") # è®“å®ŒæˆæŒ‰éˆ•é¡¯ç¤º

def pause_pomodoro():
    """æš«åœç•ªèŒ„é˜"""
    if timer_state["running"]:
        timer_state["paused"] = not timer_state["paused"]
        if timer_state["paused"]:
            timer_state["interruptions"] += 1 # æ¯æ¬¡æš«åœæ™‚ï¼Œä¸­æ–·æ¬¡æ•¸+1
            status_text = "å·²æš«åœ"
        else:
            status_text = "åŸ·è¡Œä¸­"

def end_pomodoro():
    """çµæŸç•ªèŒ„é˜ (æ”¹ç‚º generator ä»¥æä¾›å³æ™‚åé¥‹)"""
    if not timer_state["running"]:
        # ä¿æŒå›å‚³ 5 å€‹å€¼
        main_df, ai_df = update_and_get_tables()
        # ä½¿ç”¨ yield ä¾†åœæ­¢
        yield "æ²’æœ‰æ­£åœ¨é‹è¡Œçš„ç•ªèŒ„é˜", "ğŸ…", gr.update(visible=False), main_df, ai_df
        return # å¿…é ˆæœ‰ return ä¾†çµæŸç”Ÿæˆå™¨

    original_task = timer_state["task"]
    task_row_to_update = timer_state["task_row"]

    # 1. ã€å¢åŠ æç¤ºè¨Šæ¯ã€‘
    # ç«‹å³å›å‚³ä¸€å€‹ç‹€æ…‹ï¼Œå‘ŠçŸ¥ä½¿ç”¨è€…ç³»çµ±æ­£åœ¨é‹ä½œ
    # outputs=[pomodoro_status, timer_display, complete_btn, todo_table_main, todo_table_ai]
    # æˆ‘å€‘åªæ›´æ–° pomodoro_statusï¼Œå…¶ä»–å‚³å…¥ None
    yield f"è«‹ç¨ç­‰ç³»çµ±é‹ä½œï¼š{original_task}", None, None, None, None

    # [ æ¥ä¸‹ä¾†æ˜¯ç·©æ…¢çš„æ“ä½œ ]
    timer_state["running"] = False
    timer_state["manually_ended"] = "æ˜¯"

    total_duration_set = timer_state.get("total_seconds_set", 0)
    remaining_time = timer_state["remaining_time"]
    elapsed_seconds = total_duration_set - remaining_time

    timer_state["pomodoros_completed"] = elapsed_seconds // (25 * 60)

    # å‘¼å« gspread (ç·©æ…¢)
    _update_task_status(task_row_to_update, stats=timer_state)
    # [ ç·©æ…¢çš„æ“ä½œçµæŸ ]

    # æ¸…ç©ºç‹€æ…‹
    timer_state["task"] = ""
    timer_state["total_seconds_set"] = 0

    # [ å›å‚³æœ€çµ‚çµæœ ]
    # å–å¾—æ›´æ–°å¾Œçš„è¡¨æ ¼
    main_df, ai_df = update_and_get_tables()
    # é¡¯ç¤ºå®ŒæˆæŒ‰éˆ•ï¼Œä¸¦æ›´æ–°è¡¨æ ¼
    yield f"å·²æ‰‹å‹•çµæŸï¼š{original_task}", "ğŸ…", gr.update(visible=True, value="æ¨™ç¤ºç‚ºå·²å®Œæˆ ğŸ‘"), main_df, ai_df
```

```python
# æ–°å¢å‡½å¼ï¼Œç”¨æ–¼æ¨™ç¤ºä»»å‹™å·²å®Œæˆ
def mark_task_as_complete():
    """å°‡è¨ˆæ™‚å™¨å‰›çµæŸçš„ä»»å‹™æ¨™ç¤ºç‚ºã€Œå·²å®Œæˆã€"""
    task_row_to_mark = timer_state["task_row"] # å…ˆæŠ“å‡ºä¾†
    if task_row_to_mark != -1:
        _update_task_status(task_row_to_mark, new_status="å·²å®Œæˆ")
        timer_state["task_row"] = -1 # æ›´æ–°å®Œå†é‡ç½®
        # éš±è—æŒ‰éˆ•ä¸¦æ›´æ–°è¡¨æ ¼
        main_df, ai_df = update_and_get_tables()
        return gr.update(visible=False), main_df, ai_df
```

```python
# ========== åŠŸèƒ½ 3ï¼šAI å›é¥‹ ==========
def analyze_with_gemini():
    """ä½¿ç”¨ Gemini åˆ†æå¾…è¾¦äº‹é …"""
    if not model:
        return "âš ï¸ Gemini æ¨¡å‹æœªè¨­å®šï¼Œç„¡æ³•åˆ†æã€‚"

    try:
        df = get_all_todos()
        if df.empty or len(df) < 1:
            return "âš ï¸ æ²’æœ‰å¾…è¾¦äº‹é …å¯ä»¥åˆ†æã€‚"

        # æº–å‚™ Prompt
        prompt = f"""
        é€™æ˜¯æˆ‘ç›®å‰çš„å¾…è¾¦äº‹é …æ¸…å–®ï¼š
        {df.to_string(index=False)}

        è«‹æ ¹æ“šé€™ä»½æ¸…å–®ï¼Œä»¥ä¸€å€‹å°ˆæ¥­æ™‚é–“ç®¡ç†é¡§å•çš„è§’åº¦ï¼Œç”¨ç¹é«”ä¸­æ–‡æä¾›æˆ‘ä»¥ä¸‹å›é¥‹ï¼š
        1.  **ç¸½é«”è©•ä¼°**ï¼šå°æˆ‘ç›®å‰çš„ä»»å‹™é‡ã€é‡è¦æ€§èˆ‡æ€¥è¿«æ€§åˆ†ä½ˆåšä¸€å€‹ç°¡çŸ­çš„ç¸½çµã€‚
        2.  **å„ªå…ˆç´šå»ºè­°**ï¼šæ ¹æ“šã€Œè‰¾æ£®è±ªçŸ©é™£ã€(é‡è¦åˆç·Šæ€¥ > é‡è¦ä¸ç·Šæ€¥ > ç·Šæ€¥ä¸é‡è¦ > ä¸é‡è¦ä¸ç·Šæ€¥)ï¼Œå»ºè­°æˆ‘æ¥ä¸‹ä¾†æ‡‰è©²å„ªå…ˆè™•ç†å“ª 1-2 ä»¶äº‹æƒ…ï¼Œä¸¦èªªæ˜åŸå› ã€‚
        3.  **æ™‚é–“ç®¡ç†æŠ€å·§**ï¼šçµ¦æˆ‘ä¸€å€‹å…·é«”çš„å»ºè­°ï¼Œå¹«åŠ©æˆ‘æ›´æœ‰æ•ˆåœ°å®Œæˆé€™äº›ä»»å‹™ï¼ˆä¾‹å¦‚ï¼šæ˜¯å¦æœ‰äº›ä»»å‹™å¯ä»¥æ‰¹æ¬¡è™•ç†ï¼Ÿå“ªäº›ä»»å‹™é©åˆåœ¨ç²¾ç¥å¥½çš„æ™‚å€™åšï¼Ÿï¼‰ã€‚
        """

        response = model.generate_content(prompt, request_options={"timeout": 60})

        if hasattr(response, 'text') and response.text:
            advice = response.text
            # æ¸…ç† Markdown å­—å…ƒ
            advice = advice.replace("#", "").replace("*", "")
        else:
            advice = "âš ï¸ Gemini åˆ†æå®Œæˆï¼Œä½†æ²’æœ‰å›å‚³æ–‡å­—å…§å®¹ã€‚"

        # å¯«å…¥ Google Sheet
        tz = pytz.timezone("Asia/Taipei")
        timestamp = datetime.datetime.now(tz).strftime("%Y-%m-%d %H:%M:%S")
        ws_feedback.append_row([timestamp, advice], value_input_option="USER_ENTERED")

        return advice
    except Exception as e:
        return f"âŒ AI åˆ†ææ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}"
```

```python
# ========== åŠŸèƒ½ 4ï¼šè¦–è¦ºåŒ–çµ±è¨ˆ ==========

def sync_and_update_analytics():
    """
    å¾ Google Sheet åŒæ­¥æœ€æ–°è³‡æ–™ï¼Œç„¶å¾Œç”¢ç”Ÿä¸¦å›å‚³æ‰€æœ‰åœ–è¡¨ã€‚
    """
    print("ğŸ”„ æ­£åœ¨å¾ Google Sheet åŒæ­¥è³‡æ–™ä¸¦æ›´æ–°åœ–è¡¨...")
    sync_from_sheet()  # ç¬¬ä¸€æ­¥ï¼šå¼·åˆ¶åŒæ­¥
    return update_analytics_tab() # ç¬¬äºŒæ­¥ï¼šå‘¼å«ç¾æœ‰çš„å‡½å¼ä¾†ç”¢ç”Ÿåœ–è¡¨

def create_status_pie_chart(data_df):
    """æ ¹æ“š DataFrame ç”¢ç”Ÿç‹€æ…‹åˆ†ä½ˆçš„åœ“é¤…åœ–"""
    if data_df.empty or "ç‹€æ…‹" not in data_df.columns:
        # å³ä½¿æ²’æœ‰è³‡æ–™ï¼Œä¹Ÿå›å‚³ä¸€å€‹ç©ºçš„æ¨™é¡Œåœ–
        fig = px.pie(title="ç„¡è³‡æ–™å¯é¡¯ç¤º")
        fig.update_layout(annotations=[dict(text="ç„¡è³‡æ–™", x=0.5, y=0.5, font_size=20, showarrow=False)])
        return fig

    status_counts = data_df["ç‹€æ…‹"].value_counts().reset_index()
    status_counts.columns = ["ç‹€æ…‹", "æ•¸é‡"]

    fig = px.pie(status_counts,
                 names="ç‹€æ…‹",
                 values="æ•¸é‡",
                 title="ä»»å‹™ç‹€æ…‹åˆ†ä½ˆ",
                 color="ç‹€æ…‹",
                 color_discrete_map={ # è‡ªè¨‚é¡è‰²
                     "å·²å®Œæˆ": "#63C54D",
                     "é€²è¡Œä¸­": "#FFA500",
                     "æœªé–‹å§‹": "#FF6347"
                 })
    fig.update_layout(showlegend=True)
    return fig

def create_tag_bar_chart(data_df):
    """æ ¹æ“š DataFrame ç”¢ç”Ÿæ¨™ç±¤åˆ†ä½ˆçš„é•·æ¢åœ–"""
    if data_df.empty or "æ¨™ç±¤" not in data_df.columns:
        fig = px.bar(title="ç„¡è³‡æ–™å¯é¡¯ç¤º")
        fig.update_layout(annotations=[dict(text="ç„¡è³‡æ–™", x=0.5, y=0.5, font_size=20, showarrow=False)])
        return fig

    tag_counts = data_df["æ¨™ç±¤"].value_counts().reset_index()
    tag_counts.columns = ["æ¨™ç±¤", "æ•¸é‡"]

    fig = px.bar(tag_counts,
                 x="æ¨™ç±¤",
                 y="æ•¸é‡",
                 title="ä»»å‹™æ¨™ç±¤åˆ†ä½ˆ",
                 color="æ¨™ç±¤",
                 text_auto=True) # åœ¨é•·æ¢ä¸Šé¡¯ç¤ºæ•¸å­—
    return fig

def update_analytics_tab():
    """æº–å‚™è³‡æ–™ä¸¦å›å‚³æ‰€æœ‰åœ–è¡¨ (çµ¦ Gradio å‘¼å«)"""
    global todos_data
    if not todos_data:
        # å¦‚æœæ²’è³‡æ–™ï¼Œå›å‚³ç©ºçš„åœ–è¡¨
        empty_df = pd.DataFrame(columns=["ç‹€æ…‹", "æ¨™ç±¤"])
        return create_status_pie_chart(empty_df), create_tag_bar_chart(empty_df)

    # ä½¿ç”¨å…¨åŸŸè®Šæ•¸
    df = pd.DataFrame(todos_data)

    # ç¢ºä¿ 'ç‹€æ…‹' å’Œ 'æ¨™ç±¤' æ˜¯å­—ä¸²ï¼Œé¿å… Plotly è™•ç† NaNs æˆ–æ•¸å­—æ™‚å‡ºéŒ¯
    df["ç‹€æ…‹"] = df["ç‹€æ…‹"].astype(str).replace("nan", "æœªåˆ†é¡")
    df["æ¨™ç±¤"] = df["æ¨™ç±¤"].astype(str).replace("nan", "æœªåˆ†é¡")

    # ç”¢ç”Ÿåœ–è¡¨
    fig_status = create_status_pie_chart(df)
    fig_tag = create_tag_bar_chart(df)

    return fig_status, fig_tag

def update_all_components():
    """
    ä¸€å€‹é›†ä¸­çš„å‡½æ•¸ï¼Œç”¨æ–¼ç²å–æ‰€æœ‰æ›´æ–°çš„è³‡æ–™å’Œåœ–è¡¨ã€‚
    å®ƒå°‡å›å‚³ 7 å€‹é …ç›®ï¼š2 å€‹è¡¨æ ¼ï¼Œ2 å€‹åœ–è¡¨ï¼Œ3 å€‹ç¯©é¸å™¨é‡ç½®ç‹€æ…‹ã€‚
    """
    # 1. ç²å–æ›´æ–°å¾Œçš„è¡¨æ ¼
    main_df, ai_df = update_and_get_tables()

    # 2. ç²å–æ›´æ–°å¾Œçš„åœ–è¡¨
    fig_status, fig_tag = update_analytics_tab()

    # 3. æº–å‚™é‡ç½®ç¯©é¸å™¨
    reset_status_filter = gr.update(value="å…¨éƒ¨")
    reset_tag_filter = gr.update(value="å…¨éƒ¨")
    reset_due_date_filter = gr.update(value="å…¨éƒ¨")

    return main_df, ai_df, fig_status, fig_tag, reset_status_filter, reset_tag_filter, reset_due_date_filter
```
